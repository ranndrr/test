```python
# Import library yang dibutuhkan
import pandas as pd

# Memanggil / membaca dataset (contoh file excel)
df = pd.read_excel("/content/CAD alizadeh.xls")

# Menampilkan 5 data pertama
df.head()
```

```python
# Menampilkan seluruh informasi dataset
print("Informasi Dataset:")
df.info()

print("\nStatistik Deskriptif:")
df.describe()

print("\n5 Data Terakhir:")
df.tail()
```

```python
# Mengecek jumlah nilai kosong di setiap kolom
print("Jumlah data kosong tiap kolom:")
print(df.isnull().sum())

# Jika ingin menghapus baris yang memiliki nilai kosong
df = df.dropna()

# Atau jika ingin mengganti nilai kosong dengan rata-rata
# df = df.fillna(df.mean())
```

```python
# Mengubah semua kolom menjadi numerik
for column in df.columns:
    df[column] = pd.to_numeric(df[column], errors='coerce')

# Mengecek kembali tipe data
print(df.dtypes)
```

```python
df = df.fillna(df.mean())
```

```python
# Membuat salinan data
df_normalized = df.copy()

# Normalisasi tiap kolom
for col in df_normalized.columns:
    min_val = df_normalized[col].min()
    max_val = df_normalized[col].max()

    df_normalized[col] = (df_normalized[col] - min_val) / (max_val - min_val)

# Menampilkan hasil normalisasi
df_normalized.head()
```
